```python
#Devoir TP1
#NOM DES MEMBRES
# -Djiaha Kouega Marcy Audrey DJIM66300500
#- Diarietou Sow SOWD27329700


import array as arr
class Livre:
    #Constructeur de la class Livre
    def __init__(self, ID, titre, auteur, disponible=True):
        self.ID = ID
        self.titre = titre
        self. auteur = auteur
        self.disponible = disponible
        #Maniere dont un livre sera afficher
    def __str__(self):

        Disponibilites = "Disponible" if self.disponible else "Non disponible"# Assigne a une variable la condition pour determiner si un livre est disponible ou pas
        return f"livre(ID: {self.ID}, Titre: {self.titre}, Auteur: {self.auteur}, Disponibilites: {Disponibilites})"
class Auteur:
    # Constructeur de la class Auteur
    def __init__(self, nom, nationalités, oeuvres ):
        self.nom = nom
        self.nationalités = nationalités
        self.oeuvres = [] #Tableau vide pour stocker les differents oeuvres
class Emprunteur:
    # Constructeur de la class Emprunteur
    def __init__(self, ID,nom, livre_empruntes):
        self.ID = ID
        self.nom = nom
        self.livres_empruntes = []#Tableau vide pour stocker les differents livres empruntes
class Bibliotheque:
  # Constructeur de la class Emprunteur
  def __init__(self):
      self.auteurs = []
      self.livres = []
      self.emprunteurs =[]
      #Methode pour ajouter un livre a la bibliotheque
  def ajouter_livre(self, id, titre, auteur, disponible=True):
        # Vérifie si le livre existe déjà dans la bibliothèque par id
        for livre in self.livres:
            if livre.ID == id:
                print(f"Le livre d'id {id} se trouve deja dans la bibliotheque.")
                return
        # Crée un nouvel objet Livre et l'ajoute à la liste des livres
        nou_livre = Livre(id, titre, auteur, disponible)
        self.livres.append(nou_livre)
        print(f"Le livre '{titre}' de {auteur} a été ajouté à la bibliothèque avec succes.")
  # Fonction pour rechercher un livre
  def rechercher_livre(self, recherche):
      # Convertit la recherche en minuscules pour la rendre insensible a la casse
      recherche = recherche.lower()
      resultats = [livre for livre in self.livres if recherche in livre.titre.lower() or recherche in livre.auteur.lower()]
      # Affiche les résultats
      if resultats:
        print(f"Les Livres trouvés pour '{recherche}' sont:")
        for livre in resultats:
            print(livre)
      else:
          print(f"Aucun livre n'est trouvé pour '{recherche}'.")
      return resultats

  def emprunter_livre(self, id_livre, id_emprunteur):
      # Recherche du livre par ID
      livre = next((livre for livre in self.livres if livre.ID == id_livre), None)
      # Recherche de l'emprunteur par ID
      emprunteur = next((emprunteur for emprunteur in self.emprunteurs if emprunteur.ID == id_emprunteur), None)

      # Vérifie si le livre existe
      if livre is None:
          print(f"Livre avec l'ID {id_livre} non trouvé.")
          return "Livre non trouvé"

      # Vérifie si l'emprunteur existe
      if emprunteur is None:
          print(f"Emprunteur avec l'ID {id_emprunteur} n'existe pas.")
          return "Emprunteur n'existe pas"

      # Vérifie si le livre est disponible
      if livre.disponible:
          # Marque le livre comme non disponible et ajoute-le aux livres empruntés de l'emprunteur
          livre.disponible = False
          emprunteur.livres_empruntes.append(livre)
          print(f"Livre '{livre.titre}' emprunté par {emprunteur.nom}.")
          return None
      else:
          print(f"Le livre '{livre.titre}' n'est pas disponible pour emprunt.")
          return "Livre non disponible"


  def retourner_livre(self, id_livre, id_emprunteur):
    # Recherche le livre par son ID
    livre = next((livre for livre in self.livres if livre.ID == id_livre), None)
    # Recherche l'emprunteur par son ID
    emprunteur = next((emprunteur for emprunteur in self.emprunteurs if emprunteur.ID == id_emprunteur), None)

    # Vérifie si le livre existe
    if livre is None:
        print(f"Aucun livre trouvé avec l'ID {id_livre}.")
        return f"Aucun livre trouvé avec l'ID {id_livre}."

    # Vérifie si l'emprunteur existe
    if emprunteur is None:
        print(f"Aucun emprunteur trouvé avec l'ID {id_emprunteur}.")
        return f"Aucun emprunteur trouvé avec l'ID {id_emprunteur}."

    # Vérifie si l'emprunteur a emprunté ce livre
    if livre in emprunteur.livres_empruntes:
        # Supprime le livre de la liste des livres empruntés de l'emprunteur
        emprunteur.livres_empruntes.remove(livre)
        # Change la disponibilité du livre à `True`
        livre.disponible = True
        print(f"Le livre '{livre.titre}' a été retourné par {emprunteur.nom}.")
        return f"Le livre '{livre.titre}' a été retourné par {emprunteur.nom}."
    else:
        print(f"Le livre '{livre.titre}' n'a pas été emprunté par {emprunteur.nom}.")
        return "Le livre n'a pas été emprunté par cet emprunteur."


import unittest
class TestBibliotheque(unittest.TestCase):
    def setUp(self):
        self.bibliotheque = Bibliotheque()
        # Ajout des livres
        auteur1 = Auteur("Marcy Audrey", "Camerounaise", "Apprendre python")
        auteur2 = Auteur("Djiaha Kouega", "Ivoirienne", "Apprendre informatique")
        self.bibliotheque.ajouter_livre(1, "Apprendre python", "Marcy Audrey", disponible=True)
        self.bibliotheque.ajouter_livre(2, "Apprendre informatique", "Djiaha Kouega", disponible=True)
        # Ajout des emprunteurs
        emprunteur1 = Emprunteur(1, "Coco", [])
        emprunteur2 = Emprunteur(2, "Lyn", [])

        self.bibliotheque.emprunteurs.append(emprunteur1)
        self.bibliotheque.emprunteurs.append(emprunteur2)
    def test_ajouter_livre(self):
        print("-------------------------------------------------")
        print("Test 1:")
        print("Cas positif")
        # Cas positif : vérifier que le livre est ajouté
        self.bibliotheque.ajouter_livre(3, "NONONO", "Papa Wemba", disponible=True)
        self.assertEqual(len(self.bibliotheque.livres), 3)
        self.assertEqual(self.bibliotheque.livres[2].titre, "NONONO")
        print("Cas negatif")
        # Cas négatif : tenter d'ajouter un livre avec le même ID
        self.bibliotheque.ajouter_livre(1, "Python", "Prince")
        self.assertEqual(len(self.bibliotheque.livres), 3)  # Vérifie qu'aucun livre supplémentaire n'a été ajouté
        self.assertEqual(self.bibliotheque.livres[0].titre, "Apprendre python")
    def tearDown (self):
            # Nettoyage après chaque test
            print("Tearing down the test environment...")
            self.bibliotheque = None

    def test_rechercher_livre(self):
        print("-------------------------------------------------")
        print("Test 2:")
        print("Cas positif")
        # Cas positif : recherche par titre
        resultats = self.bibliotheque.rechercher_livre("Apprendre python")
        self.assertEqual(len(resultats)>0, 1)  # Si la taille du resultat est 1, le test reussi
        self.assertEqual(resultats[0].titre,
                         "Apprendre python")  # vérifie que le titre du premier livre dans les résultats est bien "Apprendre python"

        # Cas positif : recherche par auteur
        resultats = self.bibliotheque.rechercher_livre("Marcy Audrey")
        self.assertEqual(len(resultats)>0, 1)
        print("Cas negatif")
        # Cas négatif : recherche d'un titre qui n'existe pas
        resultats = self.bibliotheque.rechercher_livre("Balola")
        self.assertEqual(len(resultats)>0, 0)
        # Si la taille du resultat est 0( il ya pas de resultat, le test reussi

    def test_emprunter_livre(self):
        print("-------------------------------------------------")
        print("Test 3:")
        print("Cas positif")
        # Cas positif : emprunter un livre disponible
        if self.bibliotheque.livres[0].disponible:
            result = self.bibliotheque.emprunter_livre(1, 1)
            self.assertIsNone(result)  # Aucune erreur si l'emprunt est réussi
            self.assertFalse(self.bibliotheque.livres[0].disponible)
            self.assertIn(self.bibliotheque.livres[0],
                          self.bibliotheque.emprunteurs[0].livres_empruntes)

        print("Cas negatif")
        # Cas négatif : emprunter un livre déjà emprunté
        result = self.bibliotheque.emprunter_livre(1, 2)
        self.assertEqual(result, "Livre non disponible")

        # Cas négatif : emprunter un livre avec un emprunteur inexistant
        result = self.bibliotheque.emprunter_livre(2, 3)
        self.assertEqual(result, "Emprunteur n'existe pas")
    def test_retourner_livre(self):
        print("-------------------------------------------------")
        print("Test 4:")
        print("Cas positif")
        # Emprunter d'abord le livre pour pouvoir le retourner
        result_emprunt = self.bibliotheque.emprunter_livre(1, 1)
        self.assertIsNone(result_emprunt)  # Vérifie que l'emprunt s'est bien passé
        self.assertFalse(self.bibliotheque.livres[0].disponible)  # Le livre devrait être emprunté
        # Cas positif : retourner un livre emprunté
        result = self.bibliotheque.retourner_livre(1, 1)
        self.assertEqual(result, "Le livre 'Apprendre python' a été retourné par Coco.")
        self.assertTrue(self.bibliotheque.livres[0].disponible)
        print("Cas negatif")
        # Cas négatif : retourner un livre non emprunté
        result = self.bibliotheque.retourner_livre(2, 1)
        self.assertEqual(result, "Le livre n'a pas été emprunté par cet emprunteur.")

        # Cas négatif : retourner un livre avec un emprunteur inexistant
        result = self.bibliotheque.retourner_livre(1, 3)
        self.assertEqual(result, "Aucun emprunteur trouvé avec l'ID 3.")

# Appel des tests unitaires.
test_loader = unittest.TestLoader()  # Déclare une variable "Loader"
test_suite = test_loader.loadTestsFromTestCase(TestBibliotheque)  # Charge les tests
test_runner = unittest.TextTestRunner()  # Déclare une variable "Runner"
test_runner.run(test_suite)
```
